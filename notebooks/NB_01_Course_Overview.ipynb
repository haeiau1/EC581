## 2. Introduction to Financial Instruments and Trading

### 2.1 Structure of Financial Markets

**Exchanges (Organized Markets):**
- Only members can participate; explicit rules govern trading
- Collect and disseminate price information
- Post-trade risk and settlement (clearing houses)
- Standard contracts; market makers provide liquidity
- Examples: NYSE, LSE, Borsa Istanbul

**Over-the-Counter (OTC) Markets:**
- Non-standardized transactions
- No centralization; settlement risk
- Example: FX markets

**Buy Side vs Sell Side:**
- **Buy side:** Asset managers, hedge funds, pension funds
- **Sell side:** Brokers/dealers
- Flow: Investor → Broker → Exchange → Settlement (T+1, T+2)

### 2.2 Financial Instruments

- **Money markets:** Deposits, Repo, LIBOR/SOFR, Central Bank operations, T-Bills
- **Bond markets:** Government bonds, corporate bonds
- **FX markets:** Spot vs forward
- **Stock markets:** Equities, indices
- **Commodities:** Spot, forwards/futures
- **Derivatives:** Futures, options, warrants

### 2.3 How Trading Occurs

Primary markets (IPO) vs secondary markets. Buyers and sellers submit orders through brokers to centralized exchanges where orders are matched and transactions executed.

### 2.4 Order Types

- **Market order:** Buy/sell immediately at best available price
- **Limit order:** Buy/sell at a specified price or better
- **Stop order:** Triggered when market reaches a certain price; becomes market or limit order. Used for stop-loss and take-profit
- **Market-on-open (MOO) / Market-on-close (MOC):** Executed at session open/close
- **Good-till-cancelled (GTC):** Stays live until customer cancels
- **Fill-or-kill (FOK):** Limit order cancelled if not filled immediately

**Order Book Example:**

| Bid Quantity | Bid Price | Ask Price | Ask Quantity |
|:---:|:---:|:---:|:---:|
| 500 | 5.60 | 5.65 | 300 |
| 200 | 5.55 | 5.70 | 150 |
| 50 | 5.30 | 5.75 | 100 |
| 100 | 5.25 | 5.90 | 100 |

**Discussion Questions:**
- What happens if you send a **market order to buy 1000 lots**?
- What about a **limit order at 5.50**?

### 2.5 Trading Costs

- **Explicit costs:** Commissions, brokerage fees
- **Hidden costs:** Bid-ask spread
- **Market impact costs** (for large transactions)
- Can significantly affect overall returns and viability of arbitrage trades

### 2.6 Leverage

Use of borrowed capital or derivative instruments to increase potential return. Typical levels: margin ~2x, futures 10–20x, forex 50–100x.

**Buying on Margin:**
- Loan from broker
- Percentage margin = equity / total position
- Maintenance margin and margin calls

**Short Selling:**
- Sell a security not owned; borrow stock for delivery
- Close by buying back ("short covering")
- Borrowing cost; regulatory restrictions

**Futures Contracts:**
- Initial margin (3–10%)
- Daily mark-to-market
- Margin calls
- Clearing house guarantees

## 3. Market Efficiency

### 3.1 Efficient Market Hypothesis

Investors must be compensated for risk. The key question: **can investors earn returns beyond risk compensation (abnormal returns)?**

An informationally efficient market rapidly reflects all available public information in asset prices.

**Three forms:**
- **Weak form:** Prices reflect all past trading information
- **Semi-strong form:** Prices reflect all publicly available information
- **Strong form:** Prices reflect all information, including private/insider information

### 3.2 Market Anomalies

A strategy that generates abnormal returns exploits a "market anomaly."

**Ways of earning excess return:**
- **Risk premia investing:** Compensation for additional risks (e.g., government vs corporate bonds)
- **Forecasting:** Predicting future events and investing accordingly
- **Exploiting anomalies:** Pure riskless arbitrage, market frictions, mispricing of risks

## 4. Tools for Stock Market Analysis

### 4.1 Fundamental Analysis

Analysis of a company's operations to assess economic prospects. Based on financial statements: earnings, cash flow, profitability, financial leverage. Results in earnings growth projections and fair value estimates using equity valuation models (DCF, multiples).

### 4.2 Technical Analysis

Evaluates securities by analyzing statistics from trading activity (price, volume). Focuses on charts and analytical tools rather than intrinsic value.

**Key assumptions:**
- Price discounts everything
- Price changes are not random; trends can be identified

### 4.3 Sentiment Analysis

General prevailing attitude of investors regarding anticipated price development.
- Bullish vs bearish sentiment
- Often used as a contrarian indicator
- Extreme sentiment levels can predict market moves

### 4.4 Pre/Post-Event Trading

- Post-earnings announcement drift (PEAD)
- Event-driven strategies

## 5. Algorithmic Trading

### 5.1 Definition

Algorithmic trading is the systematic execution of trading orders decided by quantitative market models. Originally referred to splitting large orders to reduce market impact, now encompasses all algorithm-driven trading decisions.

### 5.2 Simple Examples

- **Day of the week effect:** Buy Tuesday close, sell Friday close
- **Sell-in-May:** Buy end of August, sell end of April
- **Weekly trend following:** Buy Monday open if stock was up last week; exit on 2% profit or 1.5% loss

### 5.3 Benefits

- Saves time (handles information overload)
- Avoids emotional trading
- Can be faster than manual trading
- Can be scientifically tested (backtesting)
- Most strategies are easily scalable

### 5.4 Does It Work?

"Can I make a **guaranteed** salary every month?" → **NO**

"Can I make a **random** salary, losing some months?" → **Probably, YES**
