## 1. Python Basics

### 1.1 Data Types and Variables

Python has several built-in data types that we will use extensively throughout this course:

- **`int`** — Integer numbers (e.g., share quantities, number of trading days)
- **`float`** — Floating-point numbers (e.g., prices, returns, volatilities)
- **`str`** — Strings (e.g., ticker symbols, dates)
- **`bool`** — Boolean values `True` / `False` (e.g., trading signals)
- **`None`** — Represents the absence of a value

Unlike languages such as C++ or Java, Python variables **do not require explicit type declarations**. The type is inferred at runtime.

```python
x = 10           # int
y = 3.14         # float
name = "GARAN"   # string
is_active = True # boolean

print(type(x), type(y), type(name), type(is_active))
print(f"Stock: {name}, Price: {y}, Active: {is_active}")
```

### 1.2 Collections: Lists, Tuples, Dictionaries

Python provides several built-in collection types that are essential for organizing financial data:

- **Lists** — Mutable ordered sequences. Use them to store collections of prices, tickers, returns, etc.
- **Tuples** — Immutable ordered sequences. Useful for fixed records like a trade (symbol, side, qty, price).
- **Dictionaries** — Key-value pairs. Perfect for mapping tickers to positions, prices, or metadata.

```python
# Lists
stocks = ["GARAN", "AKBNK", "THYAO", "EREGL"]
prices = [24.5, 18.2, 156.0, 45.3]
print("First stock:", stocks[0])
print("Last two:", stocks[-2:])
stocks.append("SISE")
print("After append:", stocks)

# Tuples (immutable)
trade = ("GARAN", "BUY", 100, 24.5)
symbol, side, qty, price = trade  # unpacking
print(f"Trade: {side} {qty} shares of {symbol} at {price}")

# Dictionaries
portfolio = {"GARAN": 100, "AKBNK": 200, "THYAO": 50}
portfolio["EREGL"] = 150
print("Portfolio:", portfolio)
print("GARAN position:", portfolio["GARAN"])
```

### 1.3 Control Flow

Python uses indentation (not braces) to define code blocks. The main control flow constructs are `if`/`elif`/`else`, `for` loops, `while` loops, and **list comprehensions** (a concise way to create lists).

```python
import numpy as np

# If-else
price = 24.5
if price > 25:
    signal = "SELL"
elif price < 20:
    signal = "BUY"
else:
    signal = "HOLD"
print(f"Price: {price} -> Signal: {signal}")

# For loop
portfolio = {"GARAN": 100, "AKBNK": 200, "THYAO": 50}
total_value = 0
for stock, qty in portfolio.items():
    total_value += qty
print(f"Total shares: {total_value}")

# While loop
np.random.seed(0)
capital = 10000
day = 0
while capital > 9000:
    daily_return = np.random.normal(0, 0.02)
    capital *= (1 + daily_return)
    day += 1
print(f"Capital dropped below 9000 after {day} days")

# List comprehension
returns = [0.02, -0.01, 0.03, -0.005, 0.015]
positive_returns = [r for r in returns if r > 0]
print("Positive returns:", positive_returns)
```

### 1.4 Functions

Functions allow you to encapsulate reusable logic. In quantitative finance, we frequently write functions for return calculations, risk metrics, signal generation, and more. Python also supports **lambda functions** for short, inline expressions.

```python
def calculate_return(price_start, price_end):
    """Calculate simple return."""
    return (price_end - price_start) / price_start

def annualized_return(daily_return, trading_days=252):
    """Annualize a daily return."""
    return (1 + daily_return) ** trading_days - 1

# Examples
simple_ret = calculate_return(100, 112)
print(f"Simple return: {simple_ret:.2%}")
print(f"Annualized: {annualized_return(0.0004):.2%}")

# Lambda functions
log_return = lambda p0, p1: np.log(p1 / p0)
print(f"Log return: {log_return(100, 105):.4f}")
```

## 2. NumPy

NumPy is the fundamental package for numerical computing in Python. It provides fast array operations essential for financial computations. Key advantages:

- **Vectorized operations** — perform calculations on entire arrays without explicit loops
- **Broadcasting** — automatically handle arrays of different shapes
- **Memory efficiency** — arrays stored as contiguous blocks of memory
- **Speed** — operations run in optimized C code under the hood

```python
import numpy as np

# Creating arrays
prices = np.array([100, 102, 98, 105, 103, 107, 110])
print("Prices:", prices)
print("Shape:", prices.shape, "Type:", prices.dtype)

# Vectorized operations (no loops needed!)
returns = np.diff(prices) / prices[:-1]
print("Simple returns:", np.round(returns, 4))

log_returns = np.diff(np.log(prices))
print("Log returns:", np.round(log_returns, 4))

# Basic statistics
print(f"\nMean return: {returns.mean():.4f}")
print(f"Std deviation: {returns.std():.4f}")
print(f"Min: {returns.min():.4f}, Max: {returns.max():.4f}")
print(f"Cumulative return: {(1 + returns).prod() - 1:.4f}")
```

### 2.1 Random Number Generation and Distributions

Random number generation is fundamental to quantitative finance — from Monte Carlo simulations to bootstrapping. NumPy provides fast generators for many distributions.

```python
np.random.seed(42)

# Simulate stock returns (normal distribution)
daily_returns = np.random.normal(loc=0.0005, scale=0.02, size=252)
print(f"Simulated {len(daily_returns)} daily returns")
print(f"Mean: {daily_returns.mean():.6f}, Std: {daily_returns.std():.4f}")

# Simulate a price path
S0 = 100  # initial price
price_path = S0 * np.exp(np.cumsum(daily_returns))
print(f"Start: {S0:.2f}, End: {price_path[-1]:.2f}")

# Other distributions
uniform_samples = np.random.uniform(0, 1, 1000)
t_samples = np.random.standard_t(df=5, size=1000)
```

### 2.2 Matrix Operations

Financial data is naturally multidimensional (multiple stocks over multiple days). NumPy's matrix operations make it easy to compute portfolio-level statistics, covariance matrices, and correlations.

```python
# Matrices
returns_matrix = np.random.normal(0, 0.02, (252, 5))  # 252 days, 5 stocks
print("Shape:", returns_matrix.shape)

# Column-wise statistics
mean_returns = returns_matrix.mean(axis=0)
std_returns = returns_matrix.std(axis=0)
print("Mean returns per stock:", np.round(mean_returns, 4))
print("Std per stock:", np.round(std_returns, 4))

# Correlation matrix
corr_matrix = np.corrcoef(returns_matrix.T)
print("\nCorrelation matrix:\n", np.round(corr_matrix, 2))
```

## 3. Pandas

Pandas is the most important library for data analysis in Python. It provides **DataFrame** and **Series** objects optimized for financial time series data. Key features:

- Labeled axes (rows and columns)

### 3.1 Series and DataFrame

A **Series** is a one-dimensional labeled array (think of it as a single column of data with an index). A **DataFrame** is a two-dimensional labeled data structure with columns of potentially different types (think of it as a spreadsheet or SQL table).

```python
import pandas as pd

# Series
prices = pd.Series([24.5, 25.1, 24.8, 25.5, 26.0], 
                    index=pd.date_range('2024-01-01', periods=5),
                    name='GARAN')
print(prices)
print(f"\nMean: {prices.mean():.2f}, Std: {prices.std():.2f}")

# DataFrame
data = {
    'GARAN': [24.5, 25.1, 24.8, 25.5, 26.0],
    'AKBNK': [18.2, 18.5, 18.0, 18.8, 19.1],
    'THYAO': [156.0, 158.5, 155.0, 160.0, 162.5]
}
df = pd.DataFrame(data, index=pd.date_range('2024-01-01', periods=5))
print(df)
print("\nInfo:")
print(df.describe())
```

### 3.2 Indexing and Slicing

Pandas provides powerful indexing capabilities:
- **`[]`** — column selection or boolean indexing
- **`.loc[]`** — label-based indexing (by row/column names)
- **`.iloc[]`** — integer-position based indexing

```python
# Column selection
print("Single column:")
print(df['GARAN'])
print("\nMultiple columns:")
print(df[['GARAN', 'AKBNK']])

# Row selection by label
print("\nRow by label:")
print(df.loc['2024-01-02'])

# Row selection by position
print("\nRows by position (1:3):")
print(df.iloc[1:3])

# Boolean indexing
print("\nRows where GARAN > 25:")
print(df[df['GARAN'] > 25])
```

### 3.3 Time Series Operations

Pandas excels at time series analysis. Key methods:
- **`.pct_change()`** — percentage change (simple returns)
- **`.shift()`** — shift data forward or backward in time
- **`.rolling()`** — rolling window calculations
- **`.resample()`** — change frequency (e.g., daily to monthly)
- **`.cumprod()`** / **`.cumsum()`** — cumulative operations

```python
# Calculate returns
simple_returns = df.pct_change()
print("Simple Returns:")
print(simple_returns)

log_returns = np.log(df / df.shift(1))
print("\nLog Returns:")
print(log_returns)

# Rolling windows
rolling_mean = df['GARAN'].rolling(window=3).mean()
rolling_std = df['GARAN'].rolling(window=3).std()
print("\nRolling Mean (3-day):", rolling_mean.values)

# Cumulative operations
cumulative_returns = (1 + simple_returns).cumprod() - 1
print("\nCumulative Returns (last row):")
print(cumulative_returns.iloc[-1])
```

### 3.5 GroupBy and Aggregation

The `groupby()` method splits data into groups, applies a function, and combines the results. This is useful for computing per-stock statistics, per-sector summaries, etc.

```python
# Create sample trade data
trades = pd.DataFrame({
    'symbol': ['GARAN','GARAN','AKBNK','AKBNK','THYAO','THYAO'],
    'side': ['BUY','SELL','BUY','SELL','BUY','SELL'],
    'quantity': [100, 100, 200, 200, 50, 50],
    'price': [24.5, 25.1, 18.2, 18.8, 156.0, 162.5]
})

# Group by symbol
print("Price stats by symbol:")
print(trades.groupby('symbol')['price'].agg(['mean', 'min', 'max']))

# Group by side
print("\nTotal quantity by side:")
print(trades.groupby('side')['quantity'].sum())
```

## 4. Matplotlib and Visualization

Matplotlib is the foundational plotting library in Python. We will use it extensively to visualize price series, return distributions, moving averages, and more.

```python
import matplotlib.pyplot as plt

# Simulate a price series
np.random.seed(42)
n_days = 500
returns = np.random.normal(0.0003, 0.015, n_days)
price = 100 * np.exp(np.cumsum(returns))
dates = pd.date_range('2023-01-01', periods=n_days, freq='B')

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# Line plot - Price
axes[0, 0].plot(dates, price, 'b-', linewidth=1)
axes[0, 0].set_title('Stock Price')
axes[0, 0].set_ylabel('Price')
axes[0, 0].grid(True, alpha=0.3)

# Histogram of returns
axes[0, 1].hist(returns, bins=50, edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Return Distribution')
axes[0, 1].set_xlabel('Daily Return')
axes[0, 1].axvline(x=0, color='r', linestyle='--')

# Bar chart - Monthly returns
# NOTE: Updated frequency 'M' -> 'ME' to fix ValueError with pandas >= 2.0
monthly_returns = pd.Series(returns, index=dates).resample('ME').sum()
colors = ['green' if r > 0 else 'red' for r in monthly_returns]
axes[1, 0].bar(range(len(monthly_returns)), monthly_returns, color=colors)
axes[1, 0].set_title('Monthly Returns')
axes[1, 0].set_ylabel('Return')

# Moving averages
ma20 = pd.Series(price).rolling(20).mean()
ma50 = pd.Series(price).rolling(50).mean()
axes[1, 1].plot(dates, price, 'grey', alpha=0.5, label='Price')
axes[1, 1].plot(dates, ma20, 'b-', label='MA(20)')
axes[1, 1].plot(dates, ma50, 'r-', label='MA(50)')
axes[1, 1].set_title('Price with Moving Averages')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Working with Financial Data

### 6.1 Returns

Returns are the fundamental building block of quantitative finance. We distinguish between:

- **Simple (arithmetic) return**: $R_t = \frac{P_t - P_{t-1}}{P_{t-1}}$
- **Log (continuously compounded) return**: $r_t = \ln\left(\frac{P_t}{P_{t-1}}\right) = \ln(1 + R_t)$

Log returns are additive over time: $r_{t:t+k} = r_t + r_{t+1} + \cdots + r_{t+k}$

```python
# Using SPY data
close = spy['Close'].squeeze()

# Simple returns
simple_ret = close.pct_change().dropna()

# Log returns
log_ret = np.log(close / close.shift(1)).dropna()

print("Simple returns (first 5):")
print(simple_ret.head())
print(f"\nDaily mean return: {simple_ret.mean():.6f}")
print(f"Daily std: {simple_ret.std():.4f}")
print(f"Annualized return: {simple_ret.mean() * 252:.4f}")
print(f"Annualized volatility: {simple_ret.std() * np.sqrt(252):.4f}")

# Verify relationship between simple and log returns
print(f"\nMax diff between log(1+r) and ln(P1/P0): {(np.log(1 + simple_ret) - log_ret).abs().max():.2e}")
```

### 6.2 Summary Statistics

When analyzing financial returns, we typically examine the first four moments of the distribution plus additional risk measures.

```python
from scipy import stats

ret = simple_ret

print("=== Summary Statistics for SPY Daily Returns ===")
print(f"Observations: {len(ret)}")
print(f"Mean:         {ret.mean():.6f}")
print(f"Median:       {ret.median():.6f}")
print(f"Std Dev:      {ret.std():.6f}")
print(f"Skewness:     {ret.skew():.4f}")
print(f"Kurtosis:     {ret.kurtosis():.4f} (excess)")
print(f"Min:          {ret.min():.4f}")
print(f"Max:          {ret.max():.4f}")
print(f"Sharpe Ratio: {ret.mean() / ret.std() * np.sqrt(252):.4f}")

# T-test: is mean return significantly different from zero?
t_stat, p_value = stats.ttest_1samp(ret, 0)
print(f"\nt-test: t={t_stat:.4f}, p-value={p_value:.4f}")
```

## 7. Statistical Analysis

### 7.1 Distributions in Python

The `scipy.stats` module provides a comprehensive set of probability distributions. Each distribution object has methods:

| Method | Description | R Equivalent |
|--------|-------------|-------------|
| `.pdf(x)` | Probability density function | `dnorm(x)` |
| `.cdf(x)` | Cumulative distribution function | `pnorm(x)` |
| `.ppf(q)` | Quantile function (inverse CDF) | `qnorm(q)` |
| `.rvs(size)` | Random variates | `rnorm(n)` |

```python
from scipy import stats

# Normal distribution
x = np.linspace(-4, 4, 100)
pdf = stats.norm.pdf(x, loc=0, scale=1)
cdf = stats.norm.cdf(x, loc=0, scale=1)

# Quantiles (equivalent to R's qnorm)
var_95 = stats.norm.ppf(0.05)  # 5th percentile (VaR)
print(f"95% VaR (z-score): {var_95:.4f}")

# Random samples
samples = stats.norm.rvs(loc=0.001, scale=0.02, size=1000)
print(f"Sample mean: {samples.mean():.6f}")

# Student's t distribution
t_pdf = stats.t.pdf(x, df=5)

# Plot comparison
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, pdf, 'b-', label='Normal(0,1)')
ax.plot(x, t_pdf, 'r--', label="Student-t(df=5)")
ax.set_title('Normal vs Student-t Distribution')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

### 7.2 Hypothesis Testing

Statistical hypothesis testing is central to evaluating trading strategies, comparing returns across portfolios, and testing market efficiency. The `scipy.stats` module provides many standard tests.

```python
# Two-sample t-test
np.random.seed(42)
x1 = np.random.normal(loc=0.001, scale=0.02, size=100)
x2 = np.random.normal(loc=0.003, scale=0.025, size=75)

t_stat, p_value = stats.ttest_ind(x1, x2, equal_var=True)
print(f"Two-sample t-test: t={t_stat:.4f}, p-value={p_value:.4f}")

# Custom t-test function (matching the course exercise)
def pooled_t_test(x1, x2):
    n1, n2 = len(x1), len(x2)
    mu1, mu2 = x1.mean(), x2.mean()
    s1, s2 = x1.std(ddof=1), x2.std(ddof=1)
    s_pooled2 = ((n1-1)*s1**2 + (n2-1)*s2**2) / ((n1-1) + (n2-1))
    t_stat = (mu1 - mu2) / np.sqrt(s_pooled2 * (1/n1 + 1/n2))
    return t_stat

print(f"Custom t-test: t={pooled_t_test(x1, x2):.4f}")
```

### 7.3 Autocorrelation Analysis

Autocorrelation measures the linear relationship between a time series and its lagged values. In finance:

- **ACF of returns** is typically close to zero (weak-form market efficiency)
- **ACF of absolute/squared returns** shows significant positive autocorrelation (volatility clustering)

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# ACF of returns and absolute returns
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(simple_ret.dropna(), lags=30, ax=axes[0], title='ACF of Returns')
plot_acf(simple_ret.dropna().abs(), lags=30, ax=axes[1], title='ACF of |Returns|')
plt.tight_layout()
plt.show()
```
